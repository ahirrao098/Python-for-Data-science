##Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=[-3,5,7]
```

```python
y=[10,2,5]
```

```python
plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(-3,15)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot')

plt.suptitle('Sales Comparison', size=20, y=1.03)

fig=plt.figure(figsize=(15,3))

plt.show()
```

```python
fig.set_size_inches(14,4)
```
