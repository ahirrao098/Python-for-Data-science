# Line

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
y = np.array([3,8,4,2,5])#linestyle or ls

plt.plot(y,linestyle='dotted',marker='o')

plt.show()
```

```python
y = np.array([3,8,4,2,5])#linestyle or ls

plt.plot(y,linestyle='dashed',marker='o')

plt.show()
```

```python
y = np.array([3,8,4,2,5])#linestyle or ls

plt.plot(y,ls='-.',marker='o')#shortcut symbol ls & '-.'

plt.show()
```

```python
y = np.array([3,8,4,2,5])#linestyle or ls

plt.plot(y,ls='-.',marker='o',color='#B22222')#shortcut symbol ls & '-.'

plt.show()#"#B22222" is 140 supported color name 
```

```python
y = np.array([3,8,4,2,5])#linestyle or ls

plt.plot(y,ls='-.',marker='o',color='k')#shortcut symbol ls & '-.'

plt.show()#"#B22222" is 140 supported color name 
```

```python
y = np.array([3,8,4,2,5])#linewidth or lw

plt.plot(y,ls='-.',marker='o',color='k',linewidth="10")#shortcut symbol lw & '-.'

plt.show()#"#B22222" is 140 supported color name 
```

```python
#Multiple line

y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

# Label


```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)#label x and y axis

plt.xlabel("average pulse")
plt.ylabel('calories burnage')

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)#Adding title

plt.xlabel("average pulse")
plt.ylabel('calories burnage')
plt.title('Sport watch data')

plt.show()
```

```python
# set font property
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("average pulse",fontdict=font1)
plt.ylabel('calories burnage',fontdict=font2)
plt.title('Sport watch data',fontdict=font1)

plt.show()
```
