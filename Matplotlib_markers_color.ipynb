```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([0,6])#ploting with line
y=np.array([0,250])

plt.plot(x,y)

plt.show()
```

```python
y=np.array([3,8,1,4,2])#default x axix or without x axis)

plt.plot(y)

plt.show()
```

```python
plt.plot(y, marker="o")

plt.show()
```

```python
plt.plot(y, marker="o")

plt.show()
```

```python
plt.plot(y, marker="*")

plt.show()
```

```python
plt.plot(y, marker="2")

plt.show()
```

```python
plt.plot(y,"o:r")
#dotted line :


plt.show()
```

```python
plt.plot(y,"o-r")

#sloide line "-"

plt.show()
```

```python
plt.plot(y,"o-.g")

#dash dotted line "-."

plt.show()
```

```python
plt.plot(y,"o-.g",ms=20)

#marker size is change by "ms " function-->markersize or ms

plt.show()
```

```python
plt.plot(y,"o-.g",ms=20,mec='r')

#marker edge color is change by "mec " function-->markeredgecolor or mec

plt.show()
```

```python
plt.plot(y,"o-g",ms=20,mec='r',mfc='b')

#marker face color is change by "mfc " function-->markerfacecolor or mfc

plt.show()
```
