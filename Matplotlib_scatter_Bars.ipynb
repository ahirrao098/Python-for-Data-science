# Scatter


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)

plt.show()
```

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)


x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)

plt.show()
```

```python
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color="r")


x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y, color = "k")

plt.show()
```

```python
x = np.array([5,7,8,6,3])
y = np.array([99,76,60,88,111])

colors=np.array(["r",'g','w','k','c'])

plt.scatter(x,y,c=colors,cmap="viridis")

plt.show()
```

```python
x = np.array([5,7,8,6,3,5,7,8,3,8,2,1,9,3,10,4])
y = np.array([99,76,60,88,111,61,88,90,95,105,80,98,83,107,68,71])

colors=np.array([0,13,27,35,45,5,6,7,8,9,10,20,30,40,50,60])

plt.scatter(x,y,c=colors,cmap="viridis")

plt.show()
```

```python
x = np.array([5,7,8,6,3,5,7,8,3,8,2,1,9,3,10,4])
y = np.array([99,76,60,88,111,61,88,90,95,105,80,98,83,107,68,71])

colors=np.array([0,13,27,35,45,5,6,7,8,9,10,20,30,40,50,60])

plt.scatter(x,y,c=colors,cmap="viridis")

plt.colorbar()
plt.show()
```

```python
x = np.array([5,7,8,6,3,5,7,8,3,8,2,1,9,3,10,4])
y = np.array([99,76,60,88,111,61,88,90,95,105,80,98,83,107,68,71])

sizes=np.array([0,13,27,35,45,50,6,7,8,9,100,20,30,400,50,60])

plt.scatter(x,y,s=sizes)


plt.show()
```

```python
x = np.array([5,7,8,6,3,5,7,8,3,8,2,1,9,3,10,4])
y = np.array([99,76,60,88,111,61,88,90,95,105,80,98,83,107,68,71])

sizes=np.array([0,13,27,35,45,50,6,7,8,9,100,20,30,400,50,60])

plt.scatter(x,y,s=sizes,alpha=0.5)


plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# Bars

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.barh(x, y)

plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```
