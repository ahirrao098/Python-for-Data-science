# Grid

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
x=np.array([2,5,8,1,5])
y=np.array([1,8,4,2,6])

plt.plot(x,y, marker='o',ls="dotted")

plt.xlabel("Average puls")

plt.ylabel('calories')

plt.title("Data")
plt.grid()
plt.show()
```

```python
x=np.array([2,5,8,1,5])
y=np.array([1,8,4,2,6])

plt.plot(x,y, marker='o',ls="dotted",lw="6.5")

plt.xlabel("Average puls")

plt.ylabel('calories')

plt.title("Data")
plt.grid(color="g",lw='1.5',ls='-.')
plt.show()
```

# Subplot

```python
#The subplots() function takes three arguments that describes the layout of the figure.

#The layout is organized in rows and columns, which are represented by the first and second argument.

#The third argument represents the index of the current plot.

#plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

#plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
```

```python
x=np.array([2,5,8,1,5])
y=np.array([1,8,4,2,6])

plt.subplot(2,1,1)

plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,1,2)

plt.plot(x,y)
plt.show()
```
