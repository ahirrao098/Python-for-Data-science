# Piechart

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","cherries","Dates"]

plt.pie(y,labels=mylabels)#labels

plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","cherries","Dates"]

plt.pie(y,labels=mylabels,startangle=180)#startangle

plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","cherries","Dates"]

myexplode=[0.1,0,0,0]#explode

plt.pie(y,labels=mylabels,startangle=180,explode=myexplode)

plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","cherries","Dates"]

myexplode=[0.1,0,0,0]#explode

plt.pie(y,labels=mylabels,startangle=180,explode=myexplode,shadow=True)

plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","cherries","Dates"]

myexplode=[0.1,0,0,0]#legends

plt.pie(y,labels=mylabels,startangle=180,explode=myexplode,shadow=True)

plt.legend(title="Fruits")
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","cherries","Dates"]

myexplode=[0.1,0,0,0]#legends
mycolors=["r",'k','c','g']

plt.pie(y,colors=mycolors,labels=mylabels,startangle=180,explode=myexplode,shadow=True)

plt.legend(title="Fruits")
plt.show()
```
